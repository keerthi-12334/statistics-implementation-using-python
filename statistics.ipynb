measures of central tendency method(1)


1.   mean
2.   median


1.   mode










```python
ages=[23,24,32,45,12,43,67,45,32,56,32,120]
```

```python
import numpy as np
print(np.mean(ages))
print(np.median(ages))
```

measures of central tendency method(2)



```python
import statistics
print(statistics.mean(ages))
print(statistics.median(ages))
print(statistics.mode(ages))
```


outlier visulization

5 number summary(IQR)


```python
dataset=sorted(ages)
```

```python
import numpy as np
q1,q3=np.percentile(dataset,[25,75])
print(q1,q3)
```

```python
#caluculate IQR
iqr=q3-q1
print(iqr)
```

```python
#lower fence and higher fence
lower_fence=q1-(1.5*iqr)
higher_fence=q3+(1.5*iqr)
print(lower_fence,higher_fence)
```

measures of dispersion


1.   variance
2.   standard deviation




```python
import statistics
statistics.variance(ages)
```

```python
np.var(ages,axis=0)
```

```python
def variance(data):
  n=len(ages)
  ## mean of the data
  mean=sum(data)/n
  ## variance
  deviation=[(x - mean) ** 2 for x in data]
  variance=sum(deviation)/(n-1)
  return variance

```

```python
variance(ages)
```

```python
import statistics
statistics.pvariance(ages)
```

standard deviation

```python
import math
math.sqrt(statistics.pvariance(ages))
```

check distribution is normal distirbution


```python
#### If you want to check whether feature is guassian or normal distributed
#### Q-Q plot
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
```

```python
## create a normal distributed data
s=np.random.normal(0.5,0.2,1000)
plot_data(s)
```
